# Returns 

## Net return equations:


Classic equation

R = (P<sub>final</sub> - P<sub>initial</sub>) / P<sub>initial</sub>

<br/>

Net return via time

R<sub>t</sub> = (P<sub>t</sub> - P<sub>t-1</sub>) / P<sub>t-1</sub>

## Simple Gross Return

1 + R<sub>t</sub> = P<sub>t</sub> / P<sub>t-1</sub>
- ie: invest $100 and get back $120 -> gross return is 1.2
- ie: invest $100 and get back $80 -> gross return is 0.8

## Log Return 
r<sub>t</sub> = log(1+R<sub>t</sub>) = log(P<sub>t</sub> / P<sub>t-1</sub>) = log(P<sub>t</sub>) - log(P<sub>t-1</sub>)

og Price: P<sub>t</sub> = log(P<sub>t</sub>)

## Cumulative Returns

**Given daily returns for a year, find the profit at the end of the year**

*Given: R<sub>1</sub>, R<sub>2</sub>, ..., R<sub>252</sub>, P<sub>0</sub>*
*Want: P<sub>252</sub>*
### *Answer: P<sub>252</sub> = P<sub>0</sub>(1+r<sub>1</sub>)(1+R<sub>2</sub>)...(1+R<sub>252</sub>)* 
<hr/>
<br/>


## Average Daily Return 

**What constant return would i have to get every day in order to achieve the same final P?**

### *Answer: P<sub>252</sub> = P<sub>0</sub>(1+R)<sup>252</sup>*

<br/>


## Return with Dividend

D<sub>t</sub> = dividend payment at time t

### 1 + R<sub>t</sub> = (P<sub>t</sub> + D<sub>t</sub>)  / P<sub>t-1</sub> 

or

### R<sub>t</sub> = ((P<sub>t</sub> + D<sub>t</sub>) / P<sub>t-1</sub>) - 1
